```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def F(n):
    if n<3:
        return 0
    else:
        return ((n-2)*180)/n
```

```python
F(6)
```

```python
F(60)
```

```python
def generate_positive_integers(n):
  x = []
  for i in range(1, n + 1):
    if i > 0:
      x.append(i)

  return x


if __name__ == "__main__":
  x = generate_positive_integers(100)
  print(x)

y=[]
for i in range(len(x)):
  y.append(F(x[i]))
```

```python
plt.plot(x,y,color='red',linestyle='-')
plt.title('Measure of Each Interior Angle of Polygons w.r.t their Number of Sides')
plt.xlabel('Number of Sides in The Polygon')
plt.ylabel('Measure of Each Interior Angle')
```

```python
def f(n):
    if n<3:
        return 0
    else:
        return 360/n
```

```python
f(4)
```

```python
z=[]
for i in range(len(x)):
  z.append(f(x[i]))
```

```python
plt.plot(x,z,color='red',linestyle='-')
plt.title('Measure of Each Exterior Angle of Polygons w.r.t their Number of Sides')
plt.xlabel('Number of Sides in The Polygon')
plt.ylabel('Measure of Each Exterior Angle')
```

```python
def D(n):
    if n<=3:
        return 0
    else:
        return (n*(n-3))/2
```

```python
p=[]
for i in range(len(x)):
  p.append(D(x[i]))
```

```python
plt.plot(x,p,color='red',linestyle='-')
plt.title('Number of Diagonals w.r.t Side of Polygons')
plt.xlabel('Number of Sides of Polygon')
plt.ylabel('Number of Diagonals')
```
